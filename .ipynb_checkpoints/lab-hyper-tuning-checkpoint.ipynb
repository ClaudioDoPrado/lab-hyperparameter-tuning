```python
#Libraries
from sklearn.datasets import  fetch_california_housing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor,AdaBoostRegressor, GradientBoostingRegressor
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV

from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

Now perform the same as before:
- Feature Scaling
- Feature Selection


```python
spaceship.info()
```

```python
spaceship.describe()
```

```python
spaceship.shape
```

```python
spaceship.isnull().sum()
```

```python
spaceship_cleaned = spaceship.dropna()
spaceship_cleaned
```

```python
spaceship_cleaned.isnull().sum()
```

```python
spaceship_cleaned['Cabin'].head()
```

```python
spaceship_cleaned['Cabin'] = spaceship_cleaned['Cabin'].str.slice(0,1)
spaceship_cleaned.head()
```

```python
spaceship_cleaned = spaceship_cleaned.drop(['PassengerId', 'Name'], axis=1)
spaceship_cleaned.head()
```

```python
spaceship_cleaned = pd.get_dummies(spaceship_cleaned, drop_first=True)
spaceship_cleaned.head()
```

```python
X = spaceship_cleaned.drop("Transported", axis=1)
y = spaceship_cleaned["Transported"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_train_scaled.head()
```

```python
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
X_test_scaled.head()
```

- Now let's use the best model we got so far in order to see how it can improve when we fine tune it's hyperparameters.

```python
forest = RandomForestClassifier(n_estimators=100, max_depth=20)
forest.fit(X_train_scaled, y_train)
```

```python
print("Features during training:", list(X_train_scaled.columns))
```

```python
features = X_train_scaled.columns
importances = forest.feature_importances_

# Confirm the dimensions match
print("Number of features during training:", len(features))
print("Number of importance scores:", len(importances))
assert len(features) == len(importances), "Mismatch in feature count"

# Create a DataFrame or Series with correct feature names
feat_importance_df = pd.Series(importances, index=features).sort_values(ascending=False)
print(feat_importance_df)
```

```python
print("Number of features:", len(X_train_scaled))
print("Number of importance scores:", len(importances))
```

```python
# Plot feature importance
plt.figure(figsize=(8,6))
sns.barplot(x=feat_importance_df, y=feat_importance_df.index)
plt.title('Feature Importance for Predicting Calories')
plt.show()
```

- Evaluate your model

```python
pred_for = forest.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, pred_for))
print("Precision:", precision_score(y_test, pred_for, average='weighted'))
print("Recall:", recall_score(y_test, pred_for, average='weighted'))
print("F1 Score:", f1_score(y_test, pred_for, average='weighted'))
```

**Grid/Random Search**

For this lab we will use Grid Search.

```python
param_grid = {"n_estimators": np.arange(50, 501, 50),
        "max_depth": [None, 10, 20],
        "min_samples_split": [2, 5],
        "min_samples_leaf": [1, 2],
        "max_features": [None, 'sqrt', 'log2'],
        "bootstrap": [True, False],
        "criterion": ['gini', 'entropy']}
```

- Run Grid Search

```python
rf = RandomForestClassifier(random_state=42)
```

```python
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, 
                           cv=5, n_jobs=-1, verbose=2, 
                           scoring='accuracy')
```

```python
grid_search.fit(X_train_scaled, y_train)
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

- Evaluate your model

```python
best_pred = grid_search.predict(X_test_scaled)

print("Accuracy:", accuracy_score(y_test, best_pred))
print("Precision:", precision_score(y_test, best_pred, average='weighted'))
print("Recall:", recall_score(y_test, best_pred, average='weighted'))
print("F1 Score:", f1_score(y_test, best_pred, average='weighted'))
```
